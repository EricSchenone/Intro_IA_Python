```python
#Limpiar el set de datos

data.dropna() #quita los valores vacios
```

```python
data_na = data.dropna() #Guardamos en una variable el set sin valores vacios
data_na.info()
```

```python
#Convertir la caracteristica categorica a numérica
#Proximidad al océano
data_na["ocean_proximity"]
```

```python
#Dummies / One Hot Encoding
#NEAR BAY   INLAND  NEAR OCEAN
#   1         0         0
#   0         0         1
#   0         1         0

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)
```

```python
data_na = data_na.join(dummies)#unimos los datos con los dummies
```

```python
data_na.head() #verificamos que se halla unido
```

```python
#Eliminamos la columna categorica
data_na = data_na.drop(["ocean_proximity"], axis=1) #Axis=1 indica que ocean_proximity es una columna
```

```python
#Separar las carac de las etiquetas
x = data_na.drop(["median_house_value"], axis=1)
y = data_na["median_house_value"]
```

```python
#Separar los datos en 2 partes, un conjunto de entrenamiento y uno de prueba
from sklearn.model_selection import train_test_split # type: ignore

x_tr, x_test, y_tr, y_test = train_test_split(x, y, test_size=.2)
```

```python
x_tr.shape #Cantidad de registros
```

```python
from sklearn.linear_model import LinearRegression # type: ignore

model = LinearRegression()
```

```python
model.fit(x_tr, y_tr)
```

```python
predictions = model.predict(x_test)
```

```python
comparative = {"Prediction" : predictions, "Real Value": y_test }
pd.DataFrame(comparative)
```

```python
#Overfiting o sobreajuste: es cuando los modelos de precciones no funcionan correctamente. 
#Aprenden mucho de los dtos de entrenamiento pero no pueden generalizar la respuesta
#Calculamos el score de los datos de entrenamiento y los de prueba

print(model.score(x_tr, y_tr))
print(model.score(x_test,y_test))
```

```python
#Calculamos el Error
from sklearn.metrics import mean_squared_error # type: ignore
import numpy as np # type: ignore

mse = mean_squared_error(y_test, predictions)
print(mse)

rmse = np.sqrt(mse)
rmse
```
