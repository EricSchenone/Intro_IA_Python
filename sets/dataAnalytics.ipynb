```python
#Limpiar el set de datos

data.dropna() #quita los valores vacios
```

```python
data_na = data.dropna() #Guardamos en una variable el set sin valores vacios
data_na.info()
```

```python
#Convertir la caracteristica categorica a numrica
#Proximidad al ocano
data_na["ocean_proximity"]
```

```python
#Dummies / One Hot Encoding
#NEAR BAY   INLAND  NEAR OCEAN
#   1         0         0
#   0         0         1
#   0         1         0

dummies = pd.get_dummies(data_na["ocean_proximity"], dtype=int)
```

```python
data_na = data_na.join(dummies)#unimos los datos con los dummies
```

```python
data_na.head() #verificamos que se halla unido
```

```python
#Eliminamos la columna categorica
data_na = data_na.drop(["ocean_proximity"], axis=1) #Axis=1 indica que ocean_proximity es una columna
```
