```python
from sklearn.model_selection import train_test_split
x = data.data
y = data.target

x_tr, y_tr, x_test, y_test, = train_test_split(x, y, test_size=.3)
```

```python
from sklearn import svm
from sklearn import metrics
import numpy as np

# Corrige la forma de y si es necesario
y_tr = np.argmax(y_tr, axis=1)
y_test = np.argmax(y_test, axis=1)


kernels = ["linear", "rbf", "sigmoid"]
gammas = [1, .01, .001, .0001, .00001]

for kernel in kernels:
  for gamma in gammas:
    model = svm.SVC(kernel=kernel, gamma=gamma)
    model.fit(x_tr, y_tr)
    predictions = model.predict(x_test)
    print(f"{kernel}: Gamma: {gamma}: Exactitud:", metrics.accuracy_score(y_test, predictions))
    print(f"{kernel}: Gamma: {gamma}: PrecisiÃ³n:", metrics.precision_score(y_test, predictions))

```
