```python
import seaborn as sb

sb.scatterplot(x="height", y="weigth", data=datas)

```

```python
x = datas["height"]
y = datas["weigth"]

x_processed = x.values.reshape(-1,1)
y_processed = y.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_processed, y_processed)
```

```python
height =1.69
prediction = model.predict([[height]])
print(f"Para la altura: {height} mtrs., el peso promedio es: {prediction} kgs.")
```

```python
model.score(x_processed, y_processed)
```
